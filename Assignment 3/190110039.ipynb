```
import pandas as pd

#Importing Datasets

#Because originally the second row had the column names
dataset1= pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00528/dataset.csv", header=1, delimiter=";") 

dataset2= pd.read_csv("hmis-operation_minor_no_or_local_anaesthesia-2018-19-wb-Jan_to_Mar.csv")
```

```
#Importing Macrotrends Dataset. Particularly tricky to do.
import requests
url= "http://download.macrotrends.net/assets/php/stock_data_export.php?t=GOOG"
response= requests.get(url)

#Converted Raw binary data to string
raw_data= response.content.decode("utf-8") 
#Actual data started from 15th row and splitting all values by line
raw_data= raw_data.split("\n")[14:]

#Created a new file named macrotrends_data in writing mode and filled each line of it with line of raw data, converted it to proper csv format
with open("macrotrends_data.csv", "w") as file: 
  for row in raw_data:                          
    file.write(row+"\n")

#Finally importing dataset 3
dataset3= pd.read_csv("macrotrends_data.csv")
```
